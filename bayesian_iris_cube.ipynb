```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv('data/iris.csv')
```

```python
iris.head(10)
```

```python
iris.drop(columns='Id', inplace=True)
```

```python
def my_cube_func(x: pd.DataFrame) -> pd.DataFrame:
    return x * x * x
```

```python
x1 = 5
y1 = 7

x2 = -5
y2 = -7

d = (x1 - x2) ** 2 + (y1 - y2) ** 2
print(np.sqrt(d))

dsquare = (x1 ** 2 - x2 ** 2) ** 2 + (y1 ** 2 - y2 ** 2) ** 2
print(np.sqrt(dsquare))

dcube = (x1 ** 3 - x2 ** 3) ** 2 + (y1 ** 3 - y2 ** 3) ** 2
print(np.sqrt(dcube))
```

```python
test_trans = iris[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
test_trans = my_cube_func(test_trans)

print("DATA FRAME \n")
print('Dataframe \n', iris.head(5))
print('Dataframe after transforming \n', test_trans.head(5))
print('\n')

print("DESCRIBE DATA \n")
print('Dataframe \n', iris.describe())
print('Dataframe after transforming \n', test_trans.describe())
print('\n')

print("SKEW REPORT \n")
print('iris skew \n', iris.skew())
print('testdf skew \n', test_trans.skew())
print('\n')

print("KURTOSIS REPORT \n")
print('iris kurtosis \n', iris.kurtosis())
print('testdf kurtosis \n', test_trans.kurtosis())
print('\n')

print("CORRELATION REPORT \n")
print('iris corr \n', iris.corr())
print('testdf corr \n', test_trans.corr())
print('\n')
```

```python
X=processed_iris.iloc[:,0:4].values
y=processed_iris.iloc[:,4].values
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
from sklearn.metrics import accuracy_score, precision_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
from sklearn.model_selection import KFold, train_test_split
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
gaussian = GaussianNB()
gaussian.fit(X_train, y_train)

y_pred = gaussian.predict(X_test)
accuracy_nb = round(accuracy_score(y_test, y_pred) * 100, 2)
acc_gaussian = round(gaussian.score(X_train, y_train) * 100, 2)

cm = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='micro')
recall = recall_score(y_test, y_pred, average='micro')
f1 = f1_score(y_test, y_pred, average='micro')

print("Confusion matrix for Naive Bayes\n", cm)
print("accuracy_Naive Bayes: %.3f" % accuracy)
print("precision_Naive Bayes: %.3f" % precision)
print("recall_Naive Bayes: %.3f" % recall)
print("f1-score_Naive Bayes: %.3f" % f1)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

y_pred = gaussian.predict(X_test)
accuracy_knn = round(accuracy_score(y_test, y_pred) * 100, 2)
acc_knn = round(gaussian.score(X_train, y_train) * 100, 2)

cm = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='micro')
recall = recall_score(y_test, y_pred, average='micro')
f1 = f1_score(y_test, y_pred, average='micro')

print("Confusion matrix for KNN\n", cm)
print("accuracy_KNN: %.3f" % accuracy)
print("precision_KNN: %.3f" % precision)
print("recall_KNN: %.3f" % recall)
print("f1-score_KNN: %.3f" % f1)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

plt.subplots(figsize=(20,5))
a_index = list(range(1, 50))
a = pd.Series(dtype='float64')  # Khởi tạo Series đúng cách
x = range(1, 50)

for i in range(1, 50):
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(X_train, y_train)
    prediction = model.predict(X_test)
    
    # Gán lại giá trị thay vì append trực tiếp
    a = a.append(pd.Series(accuracy_score(y_test, prediction)), ignore_index=True)

plt.plot(a_index, a, marker="*")
plt.xticks(x)
plt.show()
```

```python
decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train, y_train)

y_pred = decision_tree.predict(X_test)
accuracy_dt = round(accuracy_score(y_test, y_pred) * 100, 2)
acc_decision_tree = round(decision_tree.score(X_train, y_train) * 100, 2)

cm = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average="micro")
recall = recall_score(y_test, y_pred, average="micro")
f1 = f1_score(y_test, y_pred, average="micro")

print("Confusion matrix for DecisionTree\n", cm)
print("accuracy_DecisionTree: %.3f" % accuracy)
print("precision_DecisionTree: %.3f" % precision)
print("recall_DecisionTree: %.3f" % recall)
print("f1-score_DecisionTree: %.3f" % f1)
```

Sinh viên làm tương tự cho pipeline: data -> square -> minmaxscale -> output. Sau đó, cho nhận xét.

Gợi ý: Hàm square sẽ làm khuếch đại sự khác nhau (khoảng cách) của giá trị dữ liệu cùng dấu (âm hoặc dương), nhưng vô tình làm hai giá trị trái dấu xích gần lại.

Sinh viên làm tương tự cho pipeline: data -> log -> minmaxscale -> output. Sau đó, cho nhận xét.

Gợi ý: When our original continuous data do not follow the bell curve, we can log transform this data to make it as “normal” as possible so that the statistical analysis results from this data become more valid. In other words, the log transformation reduces or removes the skewness of our original data. The important caveat here is that the original data has to follow or approximately follow a log-normal distribution. Otherwise, the log transformation won’t work.
