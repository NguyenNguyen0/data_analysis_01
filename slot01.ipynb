# DATA FRAME MANIPULATION

**LẤY CỘT TRONG DATAFRAME**
- `df[<Tên cột>]`: sử dụng 1 ngoặc để truyền tên cột và trả cột về dưới dạng 1 Series.
- `df[[<Tên cột>]]`: sử dụng 2 ngoặc để truyền 1 list tên cột và trả cột về dưới dạng 1 DataFrame.
    - cú pháp 2 dấu ngoặc thực chất là đang truyền 1 list tên cột.
**CÚ PHÁP CHUNG**
- `df[<Tên cột>]`: với `<Tên cột>` là 1 `string` hoặc 1 `list string`

```python
# trả về Series
df['M1']
```

```python
# trả về 1 DataFrame
df[['M1']]
```

```python
# tạo 1 DataFrame mới
cols_name = ['M1', 'M2', 'M3']
newdf = df[cols_name]
newdf
```

**CẮT DATAFRAME (SLICING)**
- `df[start:end]`: với `start` là vị trí bắt đầu, `end` là vị trí kết thúc **(index bắt đầu là `start` và index kết thúc là `end - 1`)**.

```python
df[2:97]
```

**SLICING BẰNG `loc`**
- sử dụng `loc` và `[<slice>, <col_name>]`
- `loc`: truy suất dựa trên _**label-based**_ (theo tên)
- `<slice>`: nhận đối tượng `slice` _(`[2:4]`)_, để lấy số dòng
- `<col_name>`: nhận 1 mảng tên các cột cần lấy, **mặc định lấy hết**

```python
# lấy dòng từ 2 đến 97 
df.loc[2:97]
```

```python
# lấy dòng từ 2 đến 97 và cột M1, M2, M3
df.loc[2:97, ['M1', 'M2', 'M3']]
```

**SLICING BẰNG `iloc`**

Sử dụng `iloc` và `[<slice>, <slice>]`
- `iloc`: truy suất dựa trên _**index-based**_ (theo chỉ mục)
- `<slice>`: nhận đối tượng `slice` _(`[2:4]`)_, lấy số dòng
- `<slice>`: nhận đối tượng `slice` _(`[2:4]`)_, lấy cột **mặc định lấy hết**
- _(`[start:end]`)_ index cuối là `end - 1  `

```python
# lấy dòng từ 2 đến 97
df.iloc[2:97]
```

```python
# lấy dòng từ 2 đến 97 và cột M1, M2, M3
# df.iloc[[1, 2, 3, 4], [4, 5, 6]]
df.iloc[2:97, 4:7]
```

Sử dụng phương thức `sort_values`

**LỌC DATAFRAME**
| Toán tử | Tác dụng |
|---------|----------|
|    &    | AND      |
|    \|   | OR       |
|    ~    | NOT      |
|    ==   | EQUALS   |
|    !=   | NOT EQUALS |
|    >    | GREATER THAN |
|    <    | LESS THAN |
|    >=   | GREATER THAN OR EQUAL TO |
|    <=   | LESS THAN OR EQUAL TO |

<!-- **Lưu ý**: nếu sử dụng toán tử `==` thì phải sài thêm `loc` -->

```python
# các hàng có M1 > 5
df[df['M1'] > 5]
```

```python
# kết hợp nhiều điều kiện
df[(df["M1"] > 6) & (df["M2"] > 5)]
```

```python
df[df['KT'] != 'B']
# df.loc[df['KT'] == 'C']
```

```python
df[(df['M1'] > 5) | (df['M2'] > 5)]
```

```python
# kết hợp nhiều điều kiện với loc
df.loc[:, ["M1", "M2", "M3", "KT"]][df["KT"] == "C"]
```

# DATA SUMMARY

**CÁC PHƯƠNG THỨC AGGREGATION TRONG PANDAS**

|Phương thức| Tác dụng|
|-----------|---------|
|count| Số lượng quan sát không phải giá trị null|
|min| Giá trị nhỏ nhất|
|max| Giá trị lớn nhất|
|mean| Giá trị trung bình của các giá trị|
|sum| Tổng các giá trị|
|std| Độ lệch chuẩn của các giá trị|
|var| Phương sai của các giá trị|
|median| Trung vị của các giá trị|
|prod| Tích của các giá trị|
|first| Giá trị đầu tiên|
|last| Giá trị cuối cùng|
|mad| Độ lệch tuyệt đối trung bình|
|mode| Mode của các giá trị|
|sem| Sai số chuẩn của giá trị trung bình|
|skew| Độ lệch của các giá trị|
|kurt| Độ nhọn của các giá trị|
|quantile| Giá trị tại phân vị đã cho|

phương thức `agg` và `aggregate` hoàn toàn giống nhau về chức năng

```python
# df.aggregate({'M1' : ['min', 'max'], 'M2' : ['median', 'mean']})
df.agg({'M1': ['min', 'max'], 'M2': ['median', 'mean']})
```

**GOM NHÓM**

Gom nhóm theo 1 cột hoặc nhiều cột

```python
# trả về 1 groupby object
# df.groupby("KT").count()
df.groupby("KT").aggregate("count")
```

```python
# trả về 1 groupby series
# df.groupby("KT")['M1'].count()
df.groupby("KT")["M1"].aggregate("count")
```

```python
# trả về 1 DataFrameGroupBy object
df.groupby("KV")[['M1', "M2"]].aggregate("mean")
```

```python
# group theo nhiều trường
df.groupby(["KV", "KT"])[["M1", "M2"]].agg(["min", "max"])
```

**PIVOT TABLE (BẢNG TRỤC)**

Cú pháp `pandas.pivot_table(<DataFrame>, values=<cột giá trị>, columns=<cột >, aggfunc=<các phương thức aggregate>)`
- `<DataFrame>`: DataFrame
- `<cột giá trị>`: các cột giá trị để tính toán trong aggregate
- `<cột tiêu đề>`: các cột được sử dụng để làm tiêu đề
- `<các phương thức aggregate>`: các phương thức aggregate để tính toán trên cột giá trị

```python
pd.pivot_table(df, values=["M1", "M2"], columns="KV", aggfunc=["min", "mean", "max"])
```

```python
pd.pivot_table(
    df, values=["M1", "M2"], columns=["KV", "KT"], aggfunc=["min", "mean", "max"]
)
```
